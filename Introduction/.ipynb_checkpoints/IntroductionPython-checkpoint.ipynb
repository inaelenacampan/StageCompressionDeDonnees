```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def phi(x):
    res = []
    
    for elem in x :
        if(elem>=0 and elem<=1):
            res.append(1)
        else:
            res.append(0)
    
    return res
```

```python
x = np.linspace(-3,3,300)
y = phi(x)
plt.plot(x,y)
plt.title("Niveau 0")
plt.grid()
plt.show()
```

```python
def psi(j,k,x):
    
    res = []
    
    for elem in x:
        
        if ((elem >= k/math.pow(2,j)) and (elem <(k + 0.5)/math.pow(2,j))):
            res.append(1)
        
        elif ((elem >= (k + 0.5)/math.pow(2,j)) and (elem <(k + 1)/math.pow(2,j))):
            res.append(-1)
            
        else :
            res.append(0)
            
    return res        
```

```python
x = np.linspace(-0.25,1.25,300)

for j in range(0, 4):
    for k in range(0, int(math.pow(2,j))):
        y = psi(j, k, x)
        plt.plot(x,y)
    plt.title("Niveau " + str(j+1))
    plt.grid()
    plt.show()
```
